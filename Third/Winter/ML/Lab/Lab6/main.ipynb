```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons


from kmeans import KMeans
from hierarchical import Hierarchical
from density_based_clustering import DBSCAN
```

```python
# 生成一些随机数据
data, _ = make_blobs(n_samples=300, centers=5, random_state=42)
```

```python
# 使用 KMeans 聚类算法
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# 可视化结果
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', alpha=0.7, edgecolors='b')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200)
plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# 生成一些随机数据
data, _ = make_blobs(n_samples=30, centers=5, random_state=42)
```

```python
# 使用 Hierarchical 聚类算法
hierarchical = Hierarchical(n_clusters=4)
labels = hierarchical.fit_predict(data)

# 获取每个样本的序号
sample_indices = np.arange(len(data))

# 绘制散点图，将形状规定为样本序号
for i, digit in enumerate(sample_indices):
    plt.scatter(data[i, 0], data[i, 1], marker=f"${digit}$", s=100, c=plt.cm.nipy_spectral(labels[i] / 10))
plt.title('Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# 绘制树状图
dendrogram = sch.dendrogram(sch.linkage(data, method='ward'))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Cluster Distance')
plt.show()
```

```python
# 生成一些随机数据
data, _ = make_moons(n_samples=300, noise=0.05, random_state=42)

# 使用自己实现的 DBSCAN 密度聚类算法
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan.fit(data)
labels = dbscan.labels_

# 可视化结果
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='tab10', alpha=0.7, edgecolors='b')
plt.title('DBSCAN Density-Based Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
